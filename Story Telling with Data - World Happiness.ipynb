```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# read world happiness report data 2019
df = pd.read_csv("https://raw.githubusercontent.com/rajeevratan84/data-analyst-bootcamp/master/worldhappiness2019.csv")
df.head()
```

```
df['Region'].value_counts()
```

```
# Grouping it by job title
plot_data = df.groupby(['Region'], as_index=False).mean()
plot_data
```

```
plot_data = df.groupby(['Region'], as_index=False).mean()
plot_data
```

```
plot_data = df.groupby(['Region'], as_index=False)['Healthy life expectancy'].mean()
plot_data
```

```
import plotly.express as px

plot_data = df.groupby(['Region'], as_index=False)['Healthy life expectancy'].mean()

fig = px.bar(plot_data, x='Region', y='Healthy life expectancy')
fig.show()
```

```
import plotly.express as px

# Grouping it by job title and country
plot_data = df.groupby(['Region', 'Country'], as_index=False)['Healthy life expectancy'].mean()

fig = px.bar(plot_data, x='Region', y='Healthy life expectancy', color='Country')
fig.show()
```

```
import plotly.express as px

plot_data = df.groupby(['Region'], as_index=False)['Healthy life expectancy'].mean()

fig = px.line(plot_data, x='Region', y='Healthy life expectancy')
fig.show()
```

```
import plotly.express as px

# Grouping it by job title and country
plot_data = df.groupby(['Region', 'Country'], as_index=False).mean()

fig = px.line(plot_data, x='Region', y='Healthy life expectancy', color='Country')
fig.show()
```

```
import plotly.express as px

# Grouping it by job title and country
plot_data = df.groupby(['Region', 'Country'], as_index=False).mean()

fig = px.line(plot_data, x='Region', y='Healthy life expectancy', color='Region')
fig.show()
```

```
import plotly.express as px

# Grouping it by job title and country
plot_data = df.groupby(['Region'], as_index=False).mean()

plot_data.index = pd.Index(['Africa',	'Asia',	'Central,Caribbean & Latin America',	'Europe',	'Middle East',	'North America',	'Oceania'],name='Region')
plot_data = plot_data.T
plot_data = plot_data.iloc[1:]
plot_data
```

```
import plotly.express as px

fig = px.line_polar(plot_data, theta='Region', r='Count', color='Country')
fig.show()
```
