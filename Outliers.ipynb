```python
import numpy as np
import matplotlib.pyplot as plt
import statistics
```

```python
data = np.random.randn(5000)*20+20
data
```

```python
def outlier(data):
  anamolies=[]
  random_data_std = statistics.stdev(data)
  random_data_mean = statistics.mean(data)
  anomaly_cut_off= random_data_std*3
  # print(random_data_std)
  # print(random_data_mean)
  # print(anomaly_cut_off)
  lower_limit=random_data_std-anomaly_cut_off
  upper_limit=random_data_std+anomaly_cut_off
  # print(lower_limit)
  # print(upper_limit)
  for i in data:
    if i>upper_limit or i<lower_limit:
      anamolies.append(i)
  return anamolies
```

```python
outlier(data)
```
